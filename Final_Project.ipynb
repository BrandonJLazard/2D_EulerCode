\begin{equation}
    \frac{\partial U_{ij}}{\partial t} + \frac{1}{V_{ij}} \left[F_{i+1/2}^{''}S_{i+1/2} - F_{i-1/2}^{''}S_{i-1/2} + F_{j+1/2}^{''}S_{j+1/2} - F_{j-1/2}^{''}S_{j-1/2} \right ] 
\end{equation}


\begin{equation}
U = 
\begin{bmatrix} 
\rho \\ \rho u \\ \rho v \\ e  
\end{bmatrix}
\end{equation}

\begin{equation}
F^{''} = 
\begin{bmatrix} 
\rho u^{'} \\ \rho u u^{'} + pn_x \\ \rho v u^{'} + pn_y \\ (e+p)u^{'}  
\end{bmatrix}
\end{equation}

\begin{equation}
    u^{'} = un_x + vn_y
\end{equation}

Initial conditions: 

$\rho = 1~kg/m^3$ 


$P = 10^5~N/m^2$


$T = \frac{P}{\rho R} = 348.43~K$


$c = \sqrt{RT\gamma} = 374~m/s$


$M = 2.9$ 


$u = Mc = 1085.08~m/s$


$e = \frac{P}{\gamma - 1} + \frac{\rho}{2}\left [u^2 + v^2 \right] = 1427400~N/m^2$


```python
x_en = engine[0]
y_en = engine[1]
dx = (x_en[0, -1] - x_en[0, 0]) / (IL)
dy = (y_en[-1, 0] - y_en[0, 0]) / (JL)


#### ENGINE INLET CONDITIONS
M = 2.9
g = 1.4
R = 287 
rho_in = 1 #kg/m^3
P_in = 10**5
T_in = P_in/(rho_in*R)
c_in = np.sqrt(R*T_in*g)
u_in = M*c_in
v_in = 1
e_in = P_in/(g-1) + (rho_in/2)*(u_in**2 + v_in**2)



######### APPLY INITIAL CONDITIONS TO ENTIRE GRID ####################
U1 = np.full((len(x_en[:, 0]), len(x_en[0])), rho_in) 
U2 = np.full((len(x_en[:, 0]), len(x_en[0])), rho_in*u_in)
U3 = np.full((len(x_en[:, 0]), len(x_en[0])), rho_in*v_in)
U4 = np.full((len(x_en[:, 0]), len(x_en[0])), e_in)



#Combine into 2 arrays for simple functionality
U = np.array([U1, U2, U3, U4])
p_new = np.copy(U[0])


#Calculate values to get initial time
rho = U[0]
p = (g-1)*(U[-1] - (0.5/U[0])*(U[1]**2 + U[2]**2))
T = p / (R*rho)
c = np.sqrt(R*g*T)


t = 0
dt = update_time(x_en, y_en, 0.4, U2/U1, U3/U1, c)
t += dt





######## ENFORCE BOUNDARY CONDITIONS BASED ON INITIAL CONDITION ########################
U = update_solidwall(U, x_en, y_en)
U = super_outlet(U)
U = super_inlet(U, M, rho_in, P_in)



counter = 0
counters = []
Zs = []
Z = 1



U_new = np.copy(U)

while Z >= 10**(-7):

    U_new = np.zeros_like(U)


    #Update interior points
    for j in range(1, JL-1):
        for i in range(1, IL-1):
            #Compute new interior value
            U_new[:, j, i] = compute_interior_SW(U, i, j, x_en, y_en, dt)


    #Update BC's with new values
    U_new = update_solidwall(U_new, x_en, y_en)
    U_new = super_outlet(U_new)
    U_new = super_inlet(U_new, M, rho_in, P_in)
    #U_new = update_solidwall_noghost(U_new, F, dt)

    #Calculate Z (used for convergence)
    p_new = (g-1)*(U_new[-1] - (0.5/U_new[0])*(U_new[1]**2 + U_new[2]**2))
    p_old = (g-1)*(U[-1] - (0.5/U[0])*(U[1]**2 + U[2]**2))
    Z = np.max((p_new - p_old) / p_old)
    
    #Calculate speed of sound to update time
    rho = U_new[0]
    T = p_new / (R*rho)
    c = np.sqrt(R*g*T)

    #Update time step
    dt = update_time(x_en, y_en, 0.4, U_new[1]/U_new[1], U_new[3]/U_new[1], c)

    #The old U is now the new U
    U = np.copy(U_new) 

    #Append and add time and iteration counter
    t += dt
    counter += 1
    counters.append(counter)
    Zs.append(Z)
    print(Z)



#calculate final flow quantites for the entire grid once converged
rho = U[0]
p = (g-1)*(U[-1] - (0.5/U[0])*(U[1]**2 + U[2]**2))
T = p / (R*rho)
c = np.sqrt(R*g*T)
u = U[1]/U[0]
v = U[2]/U[0]
M = u/c


#Plot something to see how it looks 
plt.pcolormesh(x_en, y_en, p, cmap = 'plasma')
plt.colorbar()
#plt.pcolormesh(x_en[1:JL-1, 1:IL-1], y_en[1:JL-1, 1:IL-1], p[1:JL-1, 1:IL-1], cmap = 'plasma')
#plt.colorbar()
```

```python
u_ex_bot = []
u_ex_top = []
p_ex_bot = []
p_ex_top = []
rho_ex_top = []

def u_ex_bott(x):
    R = 287
    M1 = 2.9
    rho1 = 1
    p1 = 10**(5)
    T1 = p1/(rho1*R)
    c1 = np.sqrt(R*1.4*T1)

    M3 = 1.944
    rho3 = 2.684
    p3 = 4.077*(10**(5))
    T3 = p3/(rho3*R)
    c3 = np.sqrt(R*1.4*T3)

    if x < 2.3388:
        u = M1*c1
    else:
        u = M3*c3 
    return u

def p_ex_bott(x):
    p1 = 10**(5)

    p3 = 4.077*(10**(5))

    if x < 2.3388:
        p = p1
    else:
        p = p3
    return p


def u_ex_topp(x):
    R = 287
    M1 = 2.9
    rho1 = 1
    p1 = 10**(5)
    T1 = p1/(rho1*R)
    c1 = np.sqrt(R*1.4*T1)

    M2 = 2.379
    p2 = 2.136*(10**(5))
    rho2 = 1.698
    T2 = p2/(R*rho2)
    c2 = np.sqrt(R*1.4*T2)

    M4 = 1.553
    rho4 = 4.0154
    p4 = 7.2261*(10**(5))
    T4 = p4/(rho4*R)
    c4 = np.sqrt(R*1.4*T4)

    if x < 0.5333333:
        u = M1*c1
    elif (x >= 0.533333) & (x < 3.0829):
        u = M2*c2
    else:
        u = M4*c4
    return u

def p_ex_topp(x):

    p1 = 10**(5)

    p2 = 2.136*(10**(5))

    p4 = 7.2261*(10**(5))


    if x < 0.5333333:
        p = p1
    elif (x >= 0.533333) & (x < 3.0829):
        p = p2
    else:
        p = p4
    return p

def rho_ex_topp(x):

    rho1 = 1

    rho2 = 1.698

    rho4 = 4.0154


    if x < 0.5333333:
        rho = rho1
    elif (x >= 0.533333) & (x < 3.0829):
        rho = rho2
    else:
        rho = rho4
    return rho



for i in range(len(x_en[1, :])):
    u_ex_bot.append(u_ex_bott(x_en[1,i]))
    u_ex_top.append(u_ex_topp(x_en[-2,i]))

    p_ex_bot.append(p_ex_bott(x_en[1,i]))
    p_ex_top.append(p_ex_topp(x_en[-2,i]))

    rho_ex_top.append(rho_ex_topp(x_en[-2, i]))
    
```
